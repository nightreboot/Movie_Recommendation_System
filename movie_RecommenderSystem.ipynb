```python
import numpy as np
import pandas as pd
```

```python
movie2=pd.read_csv("tmdb_5000_movies.csv(2).zip")
movie1=pd.read_csv('tmdb_5000_credits.csv.zip')
```

```python
movie=movie1.merge(movie2,on='title')
```

```python
movie.isnull().sum()
```

```python
# id
# title
# cast
# crew
# genres
# overview 
movie=movie[['id','title','overview','cast','crew','genres','keywords']]
```

```python
movie.dropna(inplace=True)
```

```python
import ast
def extress(text):
    L=[]
    for i in ast.literal_eval(text):           # covert string to object
        L.append(i['name'])
    return L
```

```python
movie['genres']=movie['genres'].apply(extress)
```

```python
movie['keywords']=movie['keywords'].apply(extress)
```

```python
movie['crew'].iloc[0]
```

```python
import ast
def extress1(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 4:         
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movie['cast']=movie['cast'].apply(extress1)
```

```python
def extress_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movie['crew']=movie['crew'].apply(extress_director)
```

```python
movie.head(2)
```

```python
movie['overview'].iloc[2]
```

```python
movie['overview']=movie['overview'].apply(lambda x: x.split())
```

```python
def remove_space(text):
    L=[]
    for i in text:
        L.append(i.replace(" ",""))
    return L    

    
```

```python
movie.columns
```

```python
movie['cast']=movie['cast'].apply(remove_space)
movie['crew']=movie['crew'].apply(remove_space)
movie['genres']=movie['genres'].apply(remove_space)
movie['keywords']=movie['keywords'].apply(remove_space)
```

```python
movie['text']=movie['overview']+movie['cast']+movie['crew']+movie['genres']+movie['keywords']
```

```python
movie['text']
```

```python
movie.drop(columns=['crew','cast','keywords','genres','overview'],inplace=True)
```

```python
movie['text']=movie['text'].apply(lambda x:" ".join(x))
```

### Stemming

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# ps.stem('loving')
movie.iloc[100]['text']
```

```python
ps.stem(movie.iloc[100]['text'])
```

```python
def word_similarity(tags):
    l=[]
    for i in tags.split():
        l.append(ps.stem(i))
    return " ".join(l)    
```

```python
movie['text']=movie['text'].apply(word_similarity)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cvector=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cvector.fit_transform(movie['text']).toarray()
```

```python
vectors
```

### let Find the distance between different five vectors

```python

from sklearn.metrics.pairwise import cosine_similarity 
similar=cosine_similarity(vectors)
```

```python
similar
```

```python
movie.head(4)
```

```python
def similar_count(movie_name):
    movie_index=movie[movie['title']==movie_name].index[0]
    distance=similar[movie_index]
    sorting=sorted(list(enumerate(distance)),reverse=True, key=lambda x:x[1])[1:6]

    for i in sorting:
        print(movie.iloc[i[0]].title)
```
